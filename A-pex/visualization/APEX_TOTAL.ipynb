```python
def df_hv(df):
    fronts = df["front"].to_list()
    HV = []
    for front in fronts:
        HV.append(front_hv(front))
    return HV
    
```

```python
def front_size(df):
    front_sizes = []
    for front in df["front"].to_list():
        front_sizes.append(len(front))
    return front_sizes
```

```python
#front must be a 2d matrix
# calculates the HV of a front
def front_hv(front):
    inver = [] # make all objective cost negative
    for row in front:
        temp_row = []
        for val in row:
            temp_row.append(-val)
    inver.append(temp_row)
    
    ref = [0] * len(front[0]) # always use reference point {0, 0, ..., 0}
    try:
        hv = pg.hypervolume(inver) #initialize hypervolume object
    except ValueError:
        h = 0
        return h
    # Sometimes the front is too small to use 
    try: 
        h = hv.compute(ref)
    except ValueError:
        h = 0
        print("-----------------------", "\n", "no good ref point")
        print(front)
    return h
```
